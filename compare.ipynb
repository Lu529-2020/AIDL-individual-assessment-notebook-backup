```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path(".").resolve()

cnn_overall_path = PROJECT_ROOT / "output" / "cnn" / "faster_rcnn_server_baseline_light_v1" / "metrics" / "cnn_overall_summary_with_f1.csv"
cnn_per_class_path = PROJECT_ROOT / "output" / "cnn" / "faster_rcnn_server_baseline_light_v1" / "metrics" / "cnn_per_class_summary_with_f1.csv"

print(cnn_overall_path)
print(cnn_per_class_path)

cnn_overall_df = pd.read_csv(cnn_overall_path)
cnn_per_class_df = pd.read_csv(cnn_per_class_path)

display(cnn_overall_df)
display(cnn_per_class_df)
```

```python
# ============================================================
# Cell: Load YOLO final test metrics
# ============================================================

yolo_overall_path = PROJECT_ROOT / "output" / "yolo" / "metrics" / "yolov8n_final_test2017_overall_metrics.csv"
yolo_per_class_path = PROJECT_ROOT / "output" / "yolo" / "metrics" / "yolov8n_final_test2017_per_class_metrics.csv"

print(yolo_overall_path)
print(yolo_per_class_path)

if not yolo_overall_path.exists():
    raise FileNotFoundError(f"YOLO overall metrics file not found: {yolo_overall_path}")

if not yolo_per_class_path.exists():
    raise FileNotFoundError(f"YOLO per-class metrics file not found: {yolo_per_class_path}")

yolo_overall_df = pd.read_csv(yolo_overall_path)
yolo_per_class_df = pd.read_csv(yolo_per_class_path)

display(yolo_overall_df)
display(yolo_per_class_df)
```

```python
# ============================================================
# Cell: Standardise overall metrics for model comparison
# Keep only metrics available for both Faster R-CNN and YOLOv8n
# ============================================================

def standardise_cnn_overall(cnn_df: pd.DataFrame):
    row = cnn_df.iloc[0]

    inference_time_sec = row.get("inference_time_per_image_sec")

    return pd.DataFrame([{
        "model_display_name": "Faster R-CNN",
        "model_name": row.get("model_name", "fasterrcnn_resnet50_fpn"),
        "experiment_name": row.get("experiment_name", "faster_rcnn_server_baseline_light_v1"),
        "data_split": "test2017",

        # Detection performance metrics
        "mAP@0.50": row.get("map_50"),
        "mAP@0.50:0.95": row.get("map_50_95"),
        "Precision": row.get("precision"),
        "Recall": row.get("recall"),
        "F1-score": row.get("f1"),

        # Engineering metrics
        "Inference time / image (s)": inference_time_sec,
        "Inference time / image (ms)": inference_time_sec * 1000 if pd.notna(inference_time_sec) else pd.NA,
        "FPS": 1 / inference_time_sec if pd.notna(inference_time_sec) and inference_time_sec > 0 else pd.NA,
        "Parameters": row.get("num_parameters"),
        "Model size (MB)": row.get("model_file_size_mb"),
        "Num test images": row.get("num_test_images"),
    }])


def standardise_yolo_overall(yolo_df: pd.DataFrame):
    row = yolo_df.iloc[0]

    # YOLO file may use either pipeline_time_per_image_sec or separate time components.
    pipeline_time_sec = row.get("pipeline_time_per_image_sec", pd.NA)

    if pd.isna(pipeline_time_sec):
        inference_ms = row.get("inference_time_per_image_ms", pd.NA)
        preprocess_ms = row.get("preprocess_time_per_image_ms", 0)
        postprocess_ms = row.get("postprocess_time_per_image_ms", 0)

        if pd.notna(inference_ms):
            pipeline_time_sec = (preprocess_ms + inference_ms + postprocess_ms) / 1000
        else:
            pipeline_time_sec = pd.NA

    return pd.DataFrame([{
        "model_display_name": "YOLOv8n",
        "model_name": row.get("model_name", "yolov8n"),
        "experiment_name": row.get("experiment_name", "yolov8n_final_test2017"),
        "data_split": row.get("data_split", "test2017"),

        # Detection performance metrics
        "mAP@0.50": row.get("map_50"),
        "mAP@0.50:0.95": row.get("map_50_95"),
        "Precision": row.get("precision"),
        "Recall": row.get("recall"),
        "F1-score": row.get("f1"),

        # Engineering metrics
        "Inference time / image (s)": pipeline_time_sec,
        "Inference time / image (ms)": pipeline_time_sec * 1000 if pd.notna(pipeline_time_sec) else pd.NA,
        "FPS": row.get(
            "fps",
            1 / pipeline_time_sec if pd.notna(pipeline_time_sec) and pipeline_time_sec > 0 else pd.NA
        ),
        "Parameters": row.get("num_parameters"),
        "Model size (MB)": row.get("model_file_size_mb"),
        "Num test images": row.get("num_test_images"),
    }])


cnn_overall_std_df = standardise_cnn_overall(cnn_overall_df)
yolo_overall_std_df = standardise_yolo_overall(yolo_overall_df)

overall_comparison_df = pd.concat(
    [cnn_overall_std_df, yolo_overall_std_df],
    ignore_index=True
)

# Keep only fields available for both models and useful for the report
overall_comparison_cols = [
    "model_display_name",
    "model_name",
    "experiment_name",
    "data_split",
    "Num test images",
    "mAP@0.50",
    "mAP@0.50:0.95",
    "Precision",
    "Recall",
    "F1-score",
    "Inference time / image (ms)",
    "FPS",
    "Parameters",
    "Model size (MB)",
]

overall_comparison_df = overall_comparison_df[overall_comparison_cols]

display(overall_comparison_df)
```

```python
# ============================================================
# Cell: Create report-ready overall comparison table
# ============================================================

report_overall_df = overall_comparison_df.copy()

numeric_cols = [
    "mAP@0.50",
    "mAP@0.50:0.95",
    "Precision",
    "Recall",
    "F1-score",
    "Inference time / image (ms)",
    "FPS",
    "Model size (MB)",
]

for col in numeric_cols:
    report_overall_df[col] = pd.to_numeric(report_overall_df[col], errors="coerce").round(4)

report_overall_df["Parameters"] = pd.to_numeric(
    report_overall_df["Parameters"],
    errors="coerce"
).astype("Int64")

display(report_overall_df)

report_overall_path = PROJECT_ROOT / "output" / "report_overall_model_comparison.csv"
report_overall_df.to_csv(report_overall_path, index=False)

print("Saved report-ready overall comparison table to:")
print(report_overall_path)
```

```python
# ============================================================
# Cell: Compare Faster R-CNN vs YOLOv8n per-class metrics
# Keep only metrics available for both models
# ============================================================

from pathlib import Path
import pandas as pd
import numpy as np

PROJECT_ROOT = Path(".").resolve()

# ------------------------------------------------------------
# Load Faster R-CNN per-class metrics
# ------------------------------------------------------------

cnn_per_class_path = (
    PROJECT_ROOT
    / "output"
    / "cnn"
    / "faster_rcnn_server_baseline_light_v1"
    / "metrics"
    / "cnn_per_class_summary_with_f1.csv"
)

# ------------------------------------------------------------
# Load YOLO per-class metrics
# ------------------------------------------------------------

yolo_per_class_path = (
    PROJECT_ROOT
    / "output"
    / "yolo"
    / "metrics"
    / "yolov8n_final_test2017_per_class_metrics.csv"
)

if not cnn_per_class_path.exists():
    raise FileNotFoundError(f"Faster R-CNN per-class metrics not found: {cnn_per_class_path}")

if not yolo_per_class_path.exists():
    raise FileNotFoundError(f"YOLO per-class metrics not found: {yolo_per_class_path}")

cnn_per_class_df = pd.read_csv(cnn_per_class_path)
yolo_per_class_df = pd.read_csv(yolo_per_class_path)

# ------------------------------------------------------------
# Standardise Faster R-CNN per-class metrics
# ------------------------------------------------------------

cnn_per_class_std_df = pd.DataFrame({
    "model_display_name": "Faster R-CNN",
    "model_name": "fasterrcnn_resnet50_fpn",
    "data_split": "test2017",

    # Faster R-CNN category_id is 1-based, convert to 0-based for alignment
    "class_id": cnn_per_class_df["category_id"].astype(int) - 1,
    "class_name": cnn_per_class_df["category_name"],

    "AP@0.50:0.95": cnn_per_class_df["ap_50_95"],
    "Precision": cnn_per_class_df["precision"],
    "Recall": cnn_per_class_df["recall"],
    "F1-score": cnn_per_class_df["f1"],
    "Support": cnn_per_class_df["tp"] + cnn_per_class_df["fn"],
})

# ------------------------------------------------------------
# Standardise YOLO per-class metrics
# ------------------------------------------------------------

yolo_per_class_std_df = pd.DataFrame({
    "model_display_name": "YOLOv8n",
    "model_name": "yolov8n",
    "data_split": "test2017",

    "class_id": yolo_per_class_df["class_id"].astype(int),
    "class_name": yolo_per_class_df["class_name"],

    "AP@0.50:0.95": yolo_per_class_df["ap_50_95"],
    "Precision": yolo_per_class_df["precision"],
    "Recall": yolo_per_class_df["recall"],
    "F1-score": yolo_per_class_df["f1"],
    "Support": yolo_per_class_df["support"],
})

# ------------------------------------------------------------
# Combine into one long-format comparison table
# ------------------------------------------------------------

per_class_comparison_df = pd.concat(
    [cnn_per_class_std_df, yolo_per_class_std_df],
    ignore_index=True
)

# Keep class order consistent
class_order = ["SH", "SP", "SC", "OP", "MB", "HB", "CS", "CFO", "BMFO"]

per_class_comparison_df["class_name"] = pd.Categorical(
    per_class_comparison_df["class_name"],
    categories=class_order,
    ordered=True
)

per_class_comparison_df = (
    per_class_comparison_df
    .sort_values(["class_name", "model_display_name"])
    .reset_index(drop=True)
)

# Round for display/report readability
display_df = per_class_comparison_df.copy()

for col in ["AP@0.50:0.95", "Precision", "Recall", "F1-score"]:
    display_df[col] = pd.to_numeric(display_df[col], errors="coerce").round(4)

display_df["Support"] = pd.to_numeric(display_df["Support"], errors="coerce").astype("Int64")

display(display_df)

# ------------------------------------------------------------
# Save long-format comparison table
# ------------------------------------------------------------

per_class_comparison_path = PROJECT_ROOT / "output" / "per_class_metrics_summary.csv"
per_class_comparison_df.to_csv(per_class_comparison_path, index=False)

print("Saved per-class comparison table to:")
print(per_class_comparison_path)
```

```python
# ============================================================
# Cell: Create report-ready wide per-class comparison table
# ============================================================

metric_cols = [
    "AP@0.50:0.95",
    "Precision",
    "Recall",
    "F1-score",
]

wide_tables = []

for metric in metric_cols:
    metric_wide_df = per_class_comparison_df.pivot(
        index=["class_id", "class_name"],
        columns="model_display_name",
        values=metric
    ).reset_index()

    metric_wide_df.columns.name = None

    metric_wide_df = metric_wide_df.rename(columns={
        "Faster R-CNN": f"Faster R-CNN {metric}",
        "YOLOv8n": f"YOLOv8n {metric}",
    })

    wide_tables.append(metric_wide_df)

# Merge all metric-specific wide tables
report_per_class_wide_df = wide_tables[0]

for next_df in wide_tables[1:]:
    report_per_class_wide_df = report_per_class_wide_df.merge(
        next_df,
        on=["class_id", "class_name"],
        how="left"
    )

# Add support once, using YOLO support because both models use the same test split
support_df = (
    per_class_comparison_df[per_class_comparison_df["model_display_name"] == "YOLOv8n"]
    [["class_id", "class_name", "Support"]]
    .rename(columns={"Support": "Support"})
)

report_per_class_wide_df = report_per_class_wide_df.merge(
    support_df,
    on=["class_id", "class_name"],
    how="left"
)

# Round numeric columns
for col in report_per_class_wide_df.columns:
    if col not in ["class_id", "class_name", "Support"]:
        report_per_class_wide_df[col] = pd.to_numeric(
            report_per_class_wide_df[col],
            errors="coerce"
        ).round(4)

report_per_class_wide_df["Support"] = pd.to_numeric(
    report_per_class_wide_df["Support"],
    errors="coerce"
).astype("Int64")

display(report_per_class_wide_df)

report_per_class_wide_path = PROJECT_ROOT / "output" / "report_per_class_model_comparison_wide.csv"
report_per_class_wide_df.to_csv(report_per_class_wide_path, index=False)

print("Saved report-ready wide per-class comparison table to:")
print(report_per_class_wide_path)
```
